```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
CPIL = pd.read_csv('/content/gdrive/MyDrive/Fall_2023/Econ_381/PSET_1/CPILFESL(1).csv')
GDPC = pd.read_csv('/content/gdrive/MyDrive/Fall_2023/Econ_381/PSET_1/GDPC1(1).csv')
PAYM = pd.read_csv('/content/gdrive/MyDrive/Fall_2023/Econ_381/PSET_1/PAYEMS(1).csv')
```

```python
CPIL_GDPC = pd.merge(CPIL, GDPC)
```

```python
df = pd.merge(CPIL_GDPC, PAYM)
```

```python
df['CPILFESL_PC1'] = df['CPILFESL_PC1'].astype(float)
df['GDPC1_PC1'] = df['GDPC1_PC1'].astype(float)
df['PAYEMS_PC1'] = df['PAYEMS_PC1'].astype(float)

```

```python
df.describe()
```

```python
cpil_mean = df['CPILFESL_PC1'].mean()
gdpc_mean = df['GDPC1_PC1'].mean()
pay_mean = df['PAYEMS_PC1'].mean()
```

```python
df['cpil_diff'] = df['CPILFESL_PC1'] - cpil_mean
df['gdpc_diff'] = df['GDPC1_PC1'] - gdpc_mean
df['pay_diff'] = df['PAYEMS_PC1'] - pay_mean
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df.dtypes
```

```python
df_2019_4 = df[(df['DATE'] > '1948-01-01') & (df['DATE'] < '2019-10-01')]


# Assuming df contains columns 'gdpc_diff', 'pay_diff', and 'DATE'

# Use lineplot to create a line plot for 'gdpc_diff'
sns.lineplot(data=df_2019_4, x='DATE', y='gdpc_diff', label='gdpc_diff')

# Use lineplot to create a line plot for 'pay_diff'
sns.lineplot(data=df_2019_4, x='DATE', y='pay_diff', label='pay_diff')

# Set plot title and labels
plt.title('Line Plot of gdpc_diff and pay_diff - 2019')
plt.xlabel('DATE')
plt.ylabel('Value')

# Show the legend
plt.legend()

# Save the figure to your Google Drive
plt.savefig('/content/gdrive/My Drive/Fall_2023/Econ_381/PSET_1/line_plot_gp_2019.png')  # Replace with your desired file path and name

# Show the plot
plt.show()

# Calculate the correlation matrix
corr_matrix = df_2019_4[['gdpc_diff', 'pay_diff']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap = 'coolwarm', fmt=".2f")

# Set the title
plt.title("Correlation Table - GDPC & PAYF - 2019")

#Save figure to google drive
plt.savefig('/content/gdrive/My Drive/Fall_2023/Econ_381/PSET_1/corr_2019.png')

# Show the plot
plt.show()

```
